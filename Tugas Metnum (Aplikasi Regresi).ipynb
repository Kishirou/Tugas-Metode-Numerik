Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Baca Dataset & Pemilihan Kolom yang Dipakai

```python
file_path = 'Student_Performance.csv'
data = pd.read_csv(file_path)
    
X1 = data['Hours Studied'].values.reshape(-1, 1)
X2 = data['Sample Question Papers Practiced'].values.reshape(-1, 1)
y = data['Performance Index'].values
```

Metode Regresi Linear

```python
model_linear_1 = LinearRegression()
model_linear_1.fit(X1, y)
y_pred_linear_1 = model_linear_1.predict(X1)
rms_error_linear_1 = np.sqrt(mean_squared_error(y, y_pred_linear_1))
```

Metode Eksponensial

```python
log_y = np.log(y)
model_expo_1 = LinearRegression()
model_expo_1.fit(X1, log_y)
y_pred_log_linear_3 = model_expo_1.predict(X1)
y_pred_exp_3 = np.exp(y_pred_log_linear_3)
rms_error_exp_3 = np.sqrt(mean_squared_error(y, y_pred_exp_3))
```

Grafik Problem 1

```python
plt.figure(figsize=(12, 6))
plt.scatter(X1, y, color='blue', label='Data Points')
plt.plot(X1, y_pred_linear_1, color='red', label='Regresi Linear')
plt.plot(X1, y_pred_exp_3, color='green', label='Regresi Eksponensial')
plt.xlabel('Durasi Waktu Belajar')
plt.ylabel('Nilai Ujian')
plt.title('Problem 1: Durasi waktu belajar (TB) terhadap nilai ujian')
plt.legend()
plt.grid(True)
plt.show()
```

Metode Regresi Linear

```python
model_linear_2 = LinearRegression()
model_linear_2.fit(X2, y)
y_pred_linear_2 = model_linear_2.predict(X2)
rms_error_linear_2 = np.sqrt(mean_squared_error(y, y_pred_linear_2))
```

Metode Eksponensial

```python
log_y_2 = np.log(y)
model_expo_2 = LinearRegression()
model_expo_2.fit(X2, log_y_2)
y_pred_log_linear_3_2 = model_expo_2.predict(X2)
y_pred_exp_3_2 = np.exp(y_pred_log_linear_3_2)
rms_error_exp_3_2 = np.sqrt(mean_squared_error(y, y_pred_exp_3_2))
```

Grafik Problem 2

```python
plt.figure(figsize=(12, 6))
plt.scatter(X2, y, color='blue', label='Data Points')
plt.plot(X2, y_pred_linear_2, color='red', label='Regresi Linear')
plt.plot(X2, y_pred_exp_3_2, color='green', label='Regresi Exponential')
plt.xlabel('Jumlah Latihan Soal')
plt.ylabel('Nilai Ujian')
plt.title('Problem 2: Jumlah latihan soal (NL) terhadap nilai ujian')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Problem 1 - RMS Error Regresi Linear : {rms_error_linear_1}")
print(f"Problem 1 - RMS Error Regresi Eksponensial : {rms_error_exp_3}")
print(f"Problem 2 - RMS Error Regresi Linear : {rms_error_linear_2}")
print(f"Problem 2 - RMS Error Regresi Eksponensial : {rms_error_exp_3_2}")
```
